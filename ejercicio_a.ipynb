```python
#ejercicio a
#ejercicio a
import numpy as np
from matplotlib import pyplot as plt

def algoritmo_lineal(semilla,a,c,m):
  entero = (a*semilla+c)%m
  r = entero/(m-1)
  return r,entero

cantidad = 100
semilla = 21
pseudoaleatorios = []
for i in range(cantidad):
  r,semilla = algoritmo_lineal(semilla=semilla,a=21,c=15,m=31)
  if r in pseudoaleatorios:
    print(i,r)
  pseudoaleatorios.append(r)
print(pseudoaleatorios)
print("Números pseudoaleatorios generados a través del tiempo:")
plt.plot(range(cantidad),pseudoaleatorios, marker="o", color="green")
plt.show()
```
